```python
import pandas as pd
import numpy as np
import random
import names
import matplotlib.pyplot as plt
```

```python
height = [random.randint(10, 100) for i in range(5)]
weight = [random.randint(10, 100) for i in range(5)]
```

```python
# Line Chart
plt.plot(height, weight) # x-axis, y-axis
```

```python
# Line Chart
plt.plot(height, weight); # x-axis, y-axis colon removes the object info
```

```python
# Line Chart
plt.title("Relationship between Height vs Weigth")
plt.plot(height, weight)
plt.xlabel("Height")
plt.ylabel("Weight");
```

```python
# Line Chart
plt.figure(figsize = (15, 5)) # Width, Height in inches
plt.title("Relationship between Height vs Weigth")
plt.plot(height, weight)
plt.xlabel("Height")
plt.ylabel("Weight");
```

# colors
